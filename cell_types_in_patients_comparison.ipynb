```python
def get_top_5_cell_types_and_others(adata):
    cell_counts = adata['Cluster'].value_counts()
    top_5_cells = cell_counts.head(6)
    rest_count = cell_counts.iloc[6:].sum()
    combined_counts = top_5_cells.append(pd.Series(rest_count, index=['Other']))
    return combined_counts
```
